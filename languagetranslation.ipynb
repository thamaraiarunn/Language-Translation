```python
# Install libraries
!pip install transformers sentencepiece

# Import libraries
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast

# Define the tokenizer and model
model_name = "facebook/mbart-large-50-many-to-many-mmt"
tokenizer = MBart50TokenizerFast.from_pretrained(model_name)
model = MBartForConditionalGeneration.from_pretrained(model_name)

# Set the target language code
tokenizer.src_lang = "en_XX"  # Source language: English
target_lang = "ko_KR"         # Target language: Korean

# Function to translate text
def translate_text_mbart(texts):
    """
    Translate a list of texts using mBART.
    """
    inputs = tokenizer(texts, return_tensors="pt", padding=True)
    generated_tokens = model.generate(**inputs, forced_bos_token_id=tokenizer.lang_code_to_id[target_lang])
    return [tokenizer.decode(t, skip_special_tokens=True) for t in generated_tokens]

# Input texts
input_texts = [
    "Hello, how are you?",
    "This is an example of language translation.",
    "I love programming!"
]

# Translate the texts
translated_output = translate_text_mbart(input_texts)

# Print results
for i, (input_text, translation) in enumerate(zip(input_texts, translated_output)):
    print(f"Input Text {i+1}: {input_text}")
    print(f"Translated Text {i+1}: {translation}")
    print()

```
